```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# إعداد الرسوم البيانية بشكل أجمل
import seaborn as sns

# Styling
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the data
df = pd.read_csv(r"F:\Data Analysis Projects\Restaurant\restaurant_data_cleaned.csv")
df.dropna(inplace=True)

# ✅ 1. Orders per Day
df['Date'] = pd.to_datetime(df['Date'])
orders_per_day = df.groupby('Date')['OrderID'].count()

plt.figure()
orders_per_day.plot(kind='line', marker='o', color='teal')
plt.title("Daily Orders")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ✅ 2. Top 10 Selling Items
top_items = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure()
sns.barplot(x=top_items.values, y=top_items.index, hue=top_items.index, palette='viridis', legend=False)
plt.title("Top 10 Selling Items")
plt.xlabel("Quantity Sold")
plt.ylabel("Item")
plt.tight_layout()
plt.show()

# ✅ 3. Top Waiters by Revenue
top_waiters = df.groupby('WaiterName')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure()
sns.barplot(x=top_waiters.values, y=top_waiters.index, hue=top_waiters.index, palette='mako', legend=False)
plt.title(" Top Waiters by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Waiter Name")
plt.tight_layout()
plt.show()

# ✅ 4. Payment Methods Distribution
payment_counts = df['PaymentMethod'].value_counts()

plt.figure()
payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title(" Payment Methods")
plt.ylabel("")
plt.tight_layout()
plt.show()

# ✅ 5. Preparation Time Distribution
plt.figure()
sns.histplot(df['PreparationTime(min)'], bins=30, kde=True, color='coral')
plt.title(" Preparation Time Distribution")
plt.xlabel("Preparation Time (minutes)")
plt.tight_layout()
plt.show()
```
